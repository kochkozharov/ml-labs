4. Напишите функцию, которая принимает 3 трехмерных вектора $x, y, z$ и вычисляет объем пирамиды, образованной этими векторами (используйте смешанное произведение векторов)

```python
def pyramid_volume(x, y, z):
    x = np.array(x)
    y = np.array(y)
    z = np.array(z)
    cross_product = np.cross(y, z)
    mixed_product = np.dot(x, cross_product)
    volume = abs(mixed_product) / 6.0
    return volume

print(pyramid_volume([1, 0, 0], [0, 1, 0], [0, 0, 1]))
```

5. Напишите функцию, которая принимает число N и строит матрицу NxN, с нулями внутри и единицами на границе.

```python
def create_boundary_matrix(N):
    matrix = np.zeros((N, N), dtype=int)
    matrix[0, :] = 1
    matrix[-1, :] = 1
    matrix[:, 0] = 1
    matrix[:, -1] = 1
    return matrix

N = 5
result = create_boundary_matrix(N)
print(result)
```

7. Как найти наиболее близкое значение (к заданному скаляру) в векторе?

```python
def find_closest_value(vector, target):
    differences = np.abs(vector - target)
    closest_index = np.argmin(differences)
    return vector[closest_index]

vector = np.array([1, 3, 5, 7, 9])
target = 10
closest_value = find_closest_value(vector, target)
print(closest_value)
```

10. Даны 2 массива Х и Y, постройте матрицу Коши $C$ ($C_{ij} = \frac{1}{x_i - y_j}$)

```python
def cauchy_matrix(X, Y):
    C = 1 / (X[:, np.newaxis] - Y)
    return C
X = np.array([4, 5, 6])
Y = np.array([1, 2, 3])

C = cauchy_matrix(X, Y)
print(C)
```
